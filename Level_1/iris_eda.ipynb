## Objective
The objective of this analysis is to explore the structure, distribution, and relationships within the Iris dataset using statistical summaries and visualizations to derive meaningful insights.


```python
import pandas as pd
import numpy as np
import  matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r"E:\SUMED_DATA\CODVEDA - INTERNSHIP\1) iris.csv")
df.head()
```

## Dataset Overview
The Iris dataset contains flower measurements and a categorical species label. This EDA focuses on understanding how feature values vary across species.


```python
df.describe()
```

### Statistical Summary
Summary statistics provide insights into feature ranges, central tendencies, and variability across measurements.


```python
df.hist(figsize=(10,8))
plt.tight_layout()
plt.show()
```

### Distribution Analysis
- Sepal length and width show moderate spread across samples.
- Petal length and width display more distinct value ranges, indicating stronger discriminatory power.


```python
plt.figure(figsize=(10,6))
sns.boxplot(data=df)
plt.xticks(rotation=45)
plt.show()
```

### Boxplot Analysis
Boxplots highlight feature spread and potential outliers. Petal features exhibit clearer separation, reinforcing their importance in classification tasks.


```python
plt.figure(figsize=(10,6))
sns.boxplot(x='species', y='petal_length', data=df)
plt.show()
```

```python
sns.boxplot(x='species', y='petal_width', data=df)
```

### Species-wise Feature Comparison
Petal length and width vary significantly across species, making them strong predictors for classification.


```python
sns.scatterplot(x='petal_length', y='petal_width', hue='species', data=df)
plt.show()
```

### Feature Relationships
A strong positive relationship is observed between petal length and petal width, with clear clustering by species.


```python
plt.figure(figsize=(8, 6))
sns.heatmap(df.select_dtypes(include='number').corr(),
            annot=True,
            cmap='coolwarm')
plt.show()
```

### Correlation Analysis
Petal features show high positive correlation, while sepal features exhibit weaker relationships. This suggests petal features are more informative.


```python
sns.pairplot(df, hue='species')
plt.show()
```

### Multivariate Relationships
Pair plots reveal clear class separability using petal measurements, confirming their effectiveness in distinguishing species.


## Key Insights
- Petal length and petal width are the most influential features for species differentiation.
- Sepal features provide limited separation between classes.
- The dataset shows minimal noise and clear structure, making it suitable for classification modeling.

These insights guide feature selection and model choice in subsequent stages.
